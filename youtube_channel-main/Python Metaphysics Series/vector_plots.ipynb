```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import PillowWriter
```

```python
phi = np.linspace(0, 2*np.pi, 100)

def l(phi):
    return (1+3/4 * np.sin(3*phi)) * np.array([np.cos(phi), np.sin(phi), np.zeros(len(phi))])
```

```python
lx, ly, lz = l(phi)
```

```python
origin = np.zeros([1,1])
V = np.array([lx, ly]).T
```

```python
def animate(i):
    ax.clear()
    ax.quiver(origin, origin, [[V[i,0]]], [[V[i,1]]], color=['r'], angles='xy',
          scale=1, scale_units='xy', label=r'$\vec{\ell}$(t'+'={:.2f})'.format(phi[i]))
    ax.plot(lx, ly, color='blue', label='Current Carrying Wire')
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.legend(fontsize=11, loc='upper left')
    ax.set_title('Arbitrary Wire in Space', fontsize=15)
    ax.set_xlabel('$x$', fontsize=15)
    ax.set_ylabel('$y$', fontsize=15)
    
    
fig, ax = plt.subplots(1,1, figsize=(7,7))
ani = animation.FuncAnimation(fig, animate, frames=100, interval=50)
ani.save('../../output/vec.gif',writer='pillow',fps=20, dpi=150)
```
