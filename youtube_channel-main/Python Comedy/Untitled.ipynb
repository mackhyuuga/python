```python
import numpy as np
import re
import html
```

# Part 1: The Opener

Open our repository of pickup lines

```python
lines = []
with open('../data/pickuplines') as f:
    for line in f:
        lines.append(line)
```

Too messy

```python
for i, line in enumerate(lines):
    if 'Flirty Texts That Never Fail to Impress' in line:
        print(i)
```

Get the lines of interest

```python
lines_of_interest = lines[140:160]
pickup_lines = []
for line in lines_of_interest:
    pickup_lines.append(html.unescape(re.split('<|>',line)[2]))
```

Get random pickup line

```python
pickup_line = np.random.choice(pickup_lines)
```

# Part 2: The Conversation

```python
import gpt_2_simple as gpt2
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='run1')
```

The response of the good lady

```python
pickup_line = 'Hello, this is me making the first move. Your turn.'
response = 'I\'ll pass'
```

What I should say

```python
gpt2.generate(sess,
              length=250,
              temperature=1,
              run_name='run1',
              nsamples=10,
              prefix = 'ROMEO: \n'+pickup_line+'\n\nJULIET: \n'+response+'\n\nROMEO:\n',
              batch_size=5,
              top_k=5
              )
```
